```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

#처음에 이 셀만 run 했을때, "train_test_split" name 오류가 났었음. All cell run 후, 오류해결됨
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=66)


training_accuracy = []
test_accuracy = []
#1에서 10까지 n_neighbors 를 적용
neighbors_settings = range(1,11)

for n_neighbors in neighbors_settings:
    #모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train,y_train)
    #훈련 세트 정확도 저장
    training_accuracy.append(clf.score(X_train,y_train))
    #일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="train accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("accuracy")
plt.xlabel("n_neighbors")
plt.legend()
#범례 표시

#오류를 최소화하려면 편향과 분산의 합이 최소가 되는 적당한 지점을 찾아야함. 밑의 그림에서는 이웃의 수 6이 적당하다고 보여짐.
```

```python
clf.fit(X_train, y_train)
#모수 추정
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf= KNeighborsClassifier(n_neighbors = 6)
clf.fit(X_train, y_train)
print("테스트 세트 정확도:{:.2f}".format(clf.score(X_test,y_test)))
#위의 결과를 바탕으로 이웃의 개수를 6개로 했을 때, 정확도가 0.94로 향상됨.
#이웃의 개수 3개였을 땐, 0.86이였음.
```

```python
k_list = range(1,101)
accuracies = []

for k in k_list:
	classifier = KNeighborsClassifier(n_neighbors = k)
	classifier.fit(X_train,y_train)
	accuracies.append(classifier.score(X_test, y_test))

#정확도 출력시, 소수점 2번째 자리까지만 출력하도록 설정
numA=max(accuracies)
print("최대 정확도:",'%.2f'%numA)

#최소정확도가 0.89로 최소 정확도도 높은 편으로 보임.
numB=min(accuracies)
print("최소 정확도:",'%.2f'%numB)

plt.plot(k_list, accuracies)
plt.xlabel("number of k")
plt.ylabel("test set accuracy")
plt.show()

#아래 그래프를 바탕으로 최대 정확도가 0.94인 것을 보아, 위에서 추정했던 이웃의 개수는 6개가 적당하다는 것이 증명됨.
```
